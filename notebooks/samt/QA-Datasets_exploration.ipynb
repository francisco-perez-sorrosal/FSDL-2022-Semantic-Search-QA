```python
import csv
import os
from jina import Document
```

```python
def index_generator():
    
    data_path = os.path.join('../datasets/FIQA/FIQA_answers.tsv')

    # Get Document and ID
    with open(data_path) as f:
        reader = csv.reader(f, delimiter='\t')
        for data in enumerate(reader):
            d = Document()
            d.tags['id'] = int(data[0])
            d.text = data[1]
            yield d
            
next(index_generator())
```
